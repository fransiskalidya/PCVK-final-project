```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
para = cv2.imread('/content/drive/MyDrive/PCVK/Tubes/02_5_0.jpg')
gray = cv2.cvtColor(para,cv2.COLOR_BGR2GRAY) # convert menjadi grayscale
gray = cv2.medianBlur(gray,5) # metode medianBlur untuk smoothing gambar ; 5 adalah ukuran kernel (harus bernilai ganjil)

# metode houghcircles untuk mendeteksi lingkaran;
# gray = input gambar ; cv2.HOUGH_GRADIENT = mendefinisikan metode deteksi 
# 1 (dp) = rasio kebalikan dari resolusi
# 120 (min_dist) = jarak minimum antara pusat yang terdeteksi
# 400 (param1) = ambang batas atas untuk detektor tepi canny internal
# 30 (param2) = ambang batas untuk deteksi pusat
# min_radius = radius minimum yang akan dideteksi
# max_radius = radius maksimum yang akan diteksi
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT,1,120, param1=400,param2=30,minRadius=5,maxRadius=80) 
circles_round = np.uint16(np.around(circles)) # pembuatan lingkaran

for i in circles_round[0, : ]:
    cv2.circle(para, (i[0],i[1]),i[2],(50,200,200),5) # pusat lingkaran
    cv2.circle(para,(i[0],i[1]),2,(255,0,0),3) # garis luar lingkaran

print(circles.shape) # menampilkan shape dari circles yang di hasilkan
print(circles_round) # menampilkan pembuatan lingkaran
print(str(circles_round.shape[1]) + ' Oil Strorage.') #jumlah lingkaran yang terbuat

cv2_imshow(para)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
para = cv2.imread('/content/drive/MyDrive/PCVK/Tubes/nyobak.jpg')
gray = cv2.cvtColor(para,cv2.COLOR_BGR2GRAY) # convert menjadi grayscale
# gray = cv2.medianBlur(gray,5) # metode medianBlur untuk smoothing gambar ; 5 adalah ukuran kernel (harus bernilai ganjil)

# metode houghcircles untuk mendeteksi lingkaran;
# gray = input gambar ; cv2.HOUGH_GRADIENT = mendefinisikan metode deteksi 
# 1 (dp) = rasio kebalikan dari resolusi
# 120 (min_dist) = jarak minimum antara pusat yang terdeteksi
# 400 (param1) = ambang batas atas untuk detektor tepi canny internal
# 30 (param2) = ambang batas untuk deteksi pusat
# min_radius = radius minimum yang akan dideteksi
# max_radius = radius maksimum yang akan diteksi

edges = cv2.Canny(gray,100,200)
plt.subplot(121),plt.imshow(gray,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()

circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT,1,80, param1=100,param2=30,minRadius=30,maxRadius=45) 
circles_round = np.uint16(np.around(circles)) # pembuatan lingkaran

for i in circles_round[0, : ]:
    cv2.circle(para, (i[0],i[1]),i[2],(50,200,200),5) # pusat lingkaran
    cv2.circle(para,(i[0],i[1]),2,(255,0,0),3) # garis luar lingkaran

print(circles.shape) # menampilkan shape dari circles yang di hasilkan
print(circles_round) # menampilkan pembuatan lingkaran
print(str(circles_round.shape[1]) + ' Oil Strorage.') #jumlah lingkaran yang terbuat


cv2_imshow(para)

cv2.waitKey()
cv2.destroyAllWindows()
```
